```python
from pyspark import SparkContext

# Starting SparkContext
q1 = SparkContext("local", "WordCount")

# Loading the first text file
inp = q1.textFile("file_1.txt")

# word count
c_words = inp.flatMap(lambda line: line.split()) \
                   .map(lambda word: (word.lower(), 1)) \
                   .reduceByKey(lambda a, b: a + b) \
                   .collect()

for word, count in c_words:
    print(f"{word}: {count}")

# Stop SparkContext
q1.stop()
```
